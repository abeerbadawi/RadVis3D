### 11- Bank Marketing dataset
### Labels: has the client subscribed a term deposit? (binary: 'yes','no')

```python
data= pd.read_csv('Bank_Marketing.csv',sep=';')

# We convert all the columns except the label column to a range of numbers using label encoder so we can visualize the dataset in 3D
from sklearn import preprocessing 
data1 = data.iloc[:,:-1] 
data2 = data['y']
label_encoder = preprocessing.LabelEncoder()

for f in data1.columns:
    if (data1[f].dtype == 'object'):
        data1[f]= label_encoder.fit_transform(data1[f])   
        
data1 = pd.DataFrame(data=data1)
data2 = pd.DataFrame(data=data2)
data_final = pd.concat([data2,data1],axis=1)
data_final.head()

y=data_final['y']
X=data_final.drop(['y'], axis=1)
BPs=10000
RadViz3D(y,X,BPs)
RadViz2D(y,X,BPs)
```
